Normal form game is created by passing two matrices into nash.Game()function

```python
import numpy as np 
import nashpy as nash
A=np.array([[1,-1],[-1,1]])
result=nash.Game(A)
print(result)
```

Calculate utilities

```python
A=np.array([[3,0],[5,1]])
B=np.array([[3,5],[0,1]])
sol=nash.Game(A,B)
print(sol)
```

So for every player we have two strategies

```python
# case 1: if Row player use the first strategy and the column player use the second
strategy_row=np.array([1,0])
strategy_column=np.array([0,1])
sol[strategy_row,strategy_column]
print(sol[strategy_row,strategy_column])
```

```python
# in case of incertitude 
strategy_row=np.array([1/3,2/3])
strategy_column=np.array([1,0])
sol[strategy_row,strategy_column]
print(sol[strategy_row,strategy_column])
```

Cheking which one is the best strategy 

```python
A=np.array([[1,2],[3,5]])
B=np.array([[3,5],[5,7]])
sol=nash.Game(A,B)
st_r=np.array([1,0])
st_c=np.array([1,0])
print(sol[st_r,st_c])
print(sol.is_best_response(st_r,st_c))
```

Support enumeration

```python
equilibria=sol.support_enumeration()
for eq in equilibria:
    print(eq)
```

Interpretation
row player has not to play with the first strategy anymore hes has to use the second strategy
Same think for column player

Fictitious game

```python
A = np.array([[3, 1], [0, 2]])
B = np.array([[2, 0], [1, 3]])
game=nash.Game(A,B)
```

```python
np.random.seed(0)
iterations=500
play_counts=game.fictitious_play(iterations=iterations)
for rpc,cpc in play_counts:
    print(rpc,cpc)
```
